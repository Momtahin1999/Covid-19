```python
# Find the null values
```

```python
file_path = 'application_train.csv'
```

```python
column_name = 'AMT_ANNUITY'
```

```python
df = pd.read_csv(file_path)
```

```python
null_values = df[column_name].isnull()
```

```python
print(df[null_values])
```

```python
# 12 rows contains null values.
```

```python
mean_amt_annulity = df['AMT_ANNUITY'].mean() # Get the mean value for the AMT_ANNUITY field
```

```python
print(mean_amt_annulity)
```

```python
df['AMT_ANNUITY'].fillna(mean_amt_annulity, inplace = True) # Replacing the null values with the mean value
```

```python
print(df[null_values])
```

```python
# Replaced all null values. 
```

```python
# Outliers
```

```python
from scipy.stats import zscore
```

```python
import numpy as np
```

```python
z_scores = np.abs(zscore(df[['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE','DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH']]))
```

```python
outliers = (z_scores > 3).all(axis=1)
```

```python
df_remove_outliers = df[~outliers]
```

```python
# Understanding the customers
```

```python
total_loan = df['AMT_CREDIT'].sum()
```

```python
total_defult = df[df['TARGET'] ==1]['AMT_CREDIT'].sum()
```

```python
default_percentage = (total_defult/total_loan) * 100
```

```python
age_avg = df['DAYS_BIRTH'].mean()
```

```python
avg_income = df['AMT_INCOME_TOTAL'].mean()
```

```python
education_type = df['NAME_EDUCATION_TYPE'].mode().iloc[0]
```

```python
print('Amount of loans', total_loan)
print('Amount in default', total_defult)
print('Percentage of business',default_percentage)
print('Average age',age_avg)
print('Average income', avg_income)
print('Common education', education_type)
      
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sn
```

```python
sn.set(style="whitegrid")
```

```python
labels = ['Not defualt','Default']
sizes = [100 - default_percentage, default_percentage]
colors = ['#66b3ff','#99ff99']
explode = (0,0.1)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode = explode, labels=labels, colors = colors, autopct='%1.1f%%',shadow=True,startangle=90)
plt.title('Percentage of Business in Default')
plt.show()
```

```python
fig, axes = plt.subplots(3,1, figsize=(10,15))
sn.barplot(x=df['TARGET'], y=df['DAYS_BIRTH']/-365,ax=axes[0])
axes[0].set_title('Average age by default status')
axes[0].set_ylabel('Average age')

sn.barplot(x=df['TARGET'], y=df['AMT_INCOME_TOTAL'],ax=axes[1])
axes[1].set_title('Average Income by default status')
axes[1].set_ylabel('Average income')

sn.countplot(x='NAME_EDUCATION_TYPE',data=df, hue='TARGET',ax=axes[2])
axes[2].set_title('Count of education type by default')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
